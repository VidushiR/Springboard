Part 2 ‐ Experiment and metrics design

The neighboring cities of Gotham and Metropolis have complementary circadian rhythms: on
weekdays, Ultimate Gotham is most active at night, and Ultimate Metropolis is most active
during the day. On weekends, there is reasonable activity in both cities.

However, a toll bridge, with a two way toll, between the two cities causes driver partners to tend
to be exclusive to each city. The Ultimate managers of city operations for the two cities have
proposed an experiment to encourage driver partners to be available in both cities, by
reimbursing all toll costs.


## Part 2 – Experiment and Metrics Design

### 1) What would you choose as the key measure of success of this experiment, and why?

**Chosen Metric:**  
**Percentage of drivers who complete trips in both Gotham and Metropolis during the same week.**

**Why this metric?**

- The goal of the experiment is to get drivers to work in **both cities**, not just one.
- This metric shows if drivers are actually **crossing the toll bridge** and doing trips in **both places**.
- It’s simple to measure: For each driver, check if they had trips in both cities within a week.
- It helps us compare behavior between the **group that gets toll reimbursement** and the **group that doesn’t**.
- If more drivers in the toll reimbursement group are driving in both cities, the experiment is a success.

This metric is easy to understand, directly tied to the experiment’s goal, and gives clear results to help make business decisions.

### 2) Experiment Design

#### a) How to Implement the Experiment

We propose a **randomized controlled trial**:

- **Target population:** Driver partners who operate near the toll bridge.
- **Control group:** Receives **no toll reimbursement** (business as usual).
- **Treatment group:** Receives **toll reimbursement** for any cross-city trips.

**Experimental period:**  
4 to 6 weeks, covering both weekdays and weekends to capture usage patterns across different days.

**Data to collect:**
- Driver ID
- City of each trip (Gotham or Metropolis)
- Date and time of trip
- Whether the toll bridge was crossed
- Reimbursement status



#### b) Statistical Tests

To evaluate the effect of toll reimbursement on cross-city activity:

- **Primary test:**  
  **Two-proportion z-test** or **Chi-square test**  
  > To compare the percentage of cross-city drivers in the treatment vs control group.

- **Optional (if using continuous data):**  
  **t-test** or **Mann-Whitney U test**  
  > To compare the number of cross-city trips per driver.

**Hypotheses:**
- **H₀ (null):** Toll reimbursement has no effect on driver cross-city activity.
- **H₁ (alt):** Toll reimbursement increases cross-city activity.



#### c) Interpretation and Recommendations

**If the treatment group shows a statistically significant increase in cross-city service:**
- Conclude that toll reimbursement **successfully encourages** broader driver coverage.
- Recommend **scaling the program** if cost-effective.
- Consider expanding incentives based on regional behavior patterns.

**Caveats:**
- Ensure **random assignment** is unbiased.
- Monitor for **external factors** (weather, fuel prices, city events).
- Prevent **reimbursement abuse** (e.g., unnecessary bridge crossings for compensation).
