```python
# LondonHousing
### 1. Sourcing and Loading 
#### 1.1. Importing Libraries
# Let's import the pandas, numpy libraries as pd, and np respectively. 
import pandas as pd
import numpy as np
```

```python
# Load the pyplot collection of functions from matplotlib, as plt 
import matplotlib as plt
import matplotlib.pyplot as plt
```

```python
# seaborn library for line chart 
import seaborn as sns
```

```python
#### 1.2.  Loading the data
# Make a variable called url_LondonHousePrices, and assign it the following link, enclosed in quotation-marks as a string:
url_LondonHousePrices = "https://data.london.gov.uk/download/uk-house-price-index/70ac0766-8902-4eb5-aab5-01951aaed773/UK%20House%20price%20index.xls"
```

```python
# load data from the excel file
properties = pd.read_excel(url_LondonHousePrices, sheet_name='Average price', index_col=None)
```

```python
### 2. Cleaning, transforming, and visualizing

# 2.1. Exploring dataset 
#Check the number of rows and columns in the properties DataFrame, use the .shape attribute:
properties.shape
```

```python
# 2.1 Check the first few rows of the dataset
properties.head()
```

```python
#2.1 check the datatypes 
properties.dtypes
```

```python
#2.1
# Rename the 'Unnamed: 0' column to 'date'
properties = properties.rename(columns={'Unnamed: 0': 'Borough'})

# Check the first few rows of the DataFrame after renaming the column
properties.head()
```

```python
# 2.2 Cleaning The Data
#Transpose the Data, # transposing the DataFrame, which swaps its rows and columns. 
# Assign the transposed version of the original to a new variable and call it properties_T

# Transpose to have Borough names as column headers
properties_transposed = properties.transpose()

# Reset the index to make the dates as a column and reset the column names
properties_transposed = properties_transposed.reset_index()

# Assign the values of the first row as column headings
properties_transposed.columns = properties_transposed.iloc[0]

# Drop the first row since it's now the column headers
properties_transposed = properties_transposed.drop(0)

# Check the first few rows 
properties_transposed.head()
```

```python
# 2.2 cleaning the data 
properties_transposed.columns
```

```python
# 2.3. Cleaning the data (part 2)
# Convert the index to a string and split on whitespace to extract only the date part
properties_transposed.columns = properties_transposed.columns.astype(str).str.split(' ').str[0]

properties_transposed.head()
```

```python
#2.3 Cleaning the data (part 2)
# Rename the 'NaT' column to 'ID'
properties_transposed = properties_transposed.rename(columns={'NaT': 'ID'})

properties_transposed.head()
```

```python
# 2.4.Transforming the data
# Melt the DataFrame to create a tidy format
properties_melted = pd.melt(properties_transposed, id_vars=['Borough', 'ID'], var_name='Month', value_name='Average Price')

# Convert the 'Month' column to datetime
properties_melted['Month'] = pd.to_datetime(properties_melted['Month'])

# Drop rows with null values in the 'Average Price' column
properties_melted = properties_melted.dropna(subset=['Average Price'])

# Convert the 'Average Price' column to numeric
properties_melted['Average Price'] = pd.to_numeric(properties_melted['Average Price'])

# Display the first few rows 
print(properties_melted.head())
```

```python
# 2.4 
# Check the data types of the columns
print(properties_melted.dtypes)
```

```python
# 2.5. Cleaning the data (part 3)
# Check the number of unique entries in the 'Borough' column
num_boroughs = properties_melted['Borough'].nunique()
print("Number of London Boroughs:", num_boroughs)


# Check the number of entries in the 'Borough' column
num_entries = properties_melted['Borough'].count()
print("Number of entries in the 'Borough' column:", num_entries)
```

```python
# 2.5
# Display unique entries in the 'Borough' column
unique_boroughs = properties_melted['Borough'].unique()
print("Unique Boroughs:", unique_boroughs)

# Check for duplicates in the DataFrame
duplicates = properties_melted.duplicated(subset=['Borough', 'ID', 'Month'], keep=False)
print("Number of duplicates:", duplicates.sum())
```

```python
# 2. 5
# List of valid London Boroughs
london_boroughs = [
    'City of London', 'Barking & Dagenham', 'Barnet', 'Bexley', 'Brent', 'Bromley',
    'Camden', 'Croydon', 'Ealing', 'Enfield', 'Greenwich', 'Hackney',
    'Hammersmith & Fulham', 'Haringey', 'Harrow', 'Havering', 'Hillingdon',
    'Hounslow', 'Islington', 'Kensington & Chelsea', 'Kingston upon Thames',
    'Lambeth', 'Lewisham', 'Merton', 'Newham', 'Redbridge', 'Richmond upon Thames',
    'Southwark', 'Sutton', 'Tower Hamlets', 'Waltham Forest', 'Wandsworth',
    'Westminster'
]

# Filter out non-London Boroughs
properties_filtered = properties_melted[properties_melted['Borough'].isin(london_boroughs)]
properties_filtered.head(33)
```

```python
# 2.5 filter the properties
properties_filtered['Borough'].nunique()
```

```python
# 2.6. Visualizing the data
# Select data for the 'Camden' borough
camden_data = properties_filtered[properties_filtered['Borough'] == 'Camden']

# Create a line plot of Month against Average Price
plt.figure(figsize=(10, 6))
plt.plot(camden_data['Month'], camden_data['Average Price'])
plt.xlabel('Year')
plt.ylabel('Average Price')
plt.title('Average Price of Properties in Camden Borough over Time')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

```python
# 2.6 
# Filter the data for Hackney borough
hackney_data = properties_melted[properties_melted['Borough'] == 'Hackney']

# Create the line plot
plt.figure(figsize=(10, 6))
plt.plot(hackney_data['Month'], hackney_data['Average Price'])
plt.title('Average House Price in Hackney Over Time')
plt.xlabel('Month')
plt.ylabel('Average Price')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

```python
#2.6
def plot_average_prices(df, boroughs):
    plt.figure(figsize=(12, 8))
    
    for borough in boroughs:
        df_borough = df[df['Borough'] == borough]
        sns.lineplot(x='Month', y='Average Price', data=df_borough, label=borough)
    
    plt.xlabel('Year')
    plt.ylabel('Average Price')
    plt.title('Average House Prices Over Time - Top 5 Boroughs')
    plt.legend()
    plt.xticks(rotation=45)
    plt.show()
```

```python
#2.6
top_boroughs = ['Hackney', 'Waltham Forest', 'Southwark', 'Lewisham', 'Westminster']
plot_average_prices(properties_filtered, top_boroughs)
```

```python
# 2.6
# Use lambda function to check month by month
# Extract the year from the "Month" column and store it as a new column named "Year"
properties_filtered.loc[:, 'Year'] = properties_filtered['Month'].dt.year

# Display the updated DataFrame
print(properties_filtered.head())
```

```python
# 2.6 
# Group the data by "Borough" and "Year" and calculate the average price for each group
average_prices = properties_filtered.groupby(['Borough', 'Year'])['Average Price'].mean().reset_index()

# Calculate the percentage increase in house prices over the past two decades
average_prices['Percentage Increase'] = average_prices.groupby('Borough')['Average Price'].pct_change(periods=20) * 100

# Sort the data by percentage increase in descending order
average_prices_sorted = average_prices.sort_values(by='Percentage Increase', ascending=False)

# Display the boroughs with the greatest house price increase over the past two decades
print(average_prices_sorted.head())
```

```python
# 3. # 6. Modeling Price Increase Ratio
#Modeling
def create_price_ratio(df_filter):
    # Get the average price for the years 1998 and 2018
    avg_price_1998 = df_filter[df_filter['Year'] == 1998]['Average Price'].mean()
    avg_price_2018 = df_filter[df_filter['Year'] == 2018]['Average Price'].mean()
    
    # Calculate the ratio of the average price for 1998 to 2018
    price_ratio = avg_price_1998 / avg_price_2018
    
    return price_ratio
```

```python
# 3.0
# Create an empty dictionary to store the price ratios for each borough
price_ratios = {}

# Iterate through each unique London borough
for borough in properties_filtered['Borough'].unique():
    # Get the filter for the current borough
    borough_filter = properties_filtered[properties_filtered['Borough'] == borough]
    
    # Calculate the price ratio for the current borough and store it in the dictionary
    price_ratios[borough] = create_price_ratio(borough_filter)

# Convert the dictionary to a DataFrame for better visualization
price_ratios_df = pd.DataFrame(price_ratios.items(), columns=['Borough', 'Price Ratio'])

# Sort the DataFrame by the price ratio in ascending order
price_ratios_df.sort_values(by='Price Ratio', ascending=True, inplace=True)

# Display the DataFrame with the price ratios for each borough
print(price_ratios_df)
```

```python

price_ratios_df = price_ratios_df.sort_values(by='Price Ratio', ascending=False).head(15)

# Display the top 15 boroughs with the highest price ratios
print(price_ratios_df)
```

```python
# Assuming price_ratios_df is the DataFrame sorted by 'Price Ratio'
top_15 = price_ratios_df.head(15)

# Plot the top 15 boroughs
ax = top_15[['Borough', 'Price Ratio']].plot(kind='bar')

# Set the x-axis labels as borough names
ax.set_xticklabels(top_15['Borough'], rotation=45)

# Display the plot
plt.show()
```
