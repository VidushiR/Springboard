#  Exploratory Data Analysis (EDA): Movie Recommendation System

In this notebook, we will explore the cleaned MovieLens dataset to uncover patterns in user behavior, movie popularity, rating trends, and tag/genre distributions.

We aim to build domain understanding that will guide our modeling decisions for collaborative filtering, content-based filtering, and hybrid recommendations and deep learning.


Load Data

```python
#Load the libraries

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Set up display settings
sns.set(style='whitegrid')
plt.rcParams['figure.figsize'] = (10, 6)
```

```python
#Load processed/interim data

# Load cleaned data
ratings = pd.read_csv("../data/interim/rating_cleaned.csv")
movies = pd.read_csv("../data/interim/movie_cleaned.csv")
tags = pd.read_csv("../data/interim/tag_cleaned.csv")
```

```python
#  Basic Data Overview

# Shape of each dataset
print("Movies shape:", movies.shape)
print("Ratings shape:", ratings.shape)
print("Tags shape:", tags.shape)

# Check for missing values
print("\nMissing values in movies:")
print(movies.isnull().sum())

print("\nMissing values in ratings:")
print(ratings.isnull().sum())

print("\nMissing values in tags:")
print(tags.isnull().sum())

# View first few rows
print("\nSample of movies data:")
print(movies.head())
```

```python

#Most Rated Movies
top_movies = ratings['movieId'].value_counts().head(10)
top_movie_details = movies[movies['movieId'].isin(top_movies.index)]
print(top_movie_details[['movieId', 'title']])
```

```python
# Bar Plot for Top-Rated Movies
top_movie_titles = movies[movies['movieId'].isin(top_movies.index)].copy()
top_movie_counts = ratings['movieId'].value_counts().loc[top_movie_titles['movieId']]

# Merge count
top_movie_titles.loc[:, 'rating_count'] = top_movie_titles['movieId'].map(top_movie_counts)
top_movie_titles = top_movie_titles.sort_values('rating_count', ascending=False)

plt.figure(figsize=(10, 6))
sns.barplot(data=top_movie_titles, y='title', x='rating_count')
plt.title("Top 10 Most Rated Movies")
plt.xlabel("Number of Ratings")
plt.ylabel("Movie Title")
plt.show()
```

### Top 10 Most Rated Movies

This bar plot shows the top 10 most rated movies in the dataset. These are titles that have received the highest number of ratings from users. 

Popular movies like *Pulp Fiction*, *Forrest Gump*, and *The Shawshank Redemption* dominate the top of the list.

**Key Observations:**
- The most rated movies are mostly from the 1990s, indicating high engagement with classics.
- These movies can serve as strong anchors in a collaborative filtering model since they have high visibility and user interaction.
- Including such popular titles in recommendations can improve user satisfaction, especially for new users with limited history (cold-start problem).

**Conclusion:**
This insight will help prioritize high-interaction movies in training and evaluation. It also validates the dataset's richness in user behavior and engagement trends—crucial for building robust recommendation systems.


```python
# Rating Distribution
# Distribution of Ratings
import seaborn as sns
import matplotlib.pyplot as plt

sns.histplot(data=ratings, x='rating', bins=10, kde=True, color='orange')
plt.title("Distribution of Ratings")
plt.xlabel("Rating Score")
plt.ylabel("Count")
plt.grid(True)
plt.tight_layout()
plt.show()
```

### Distribution of Movie Ratings

This histogram shows how users have rated movies in the dataset. A KDE curve has been added to understand the density of the rating values.

**Key Observations:**
- Most ratings fall between **3.0 and 5.0**, with significant peaks at **4.0** and **5.0**.
- Very few users give low ratings (below 2.0), which suggests users are more likely to rate movies they enjoyed.
- This type of distribution is typical in movie recommendation systems due to **positive user bias**.

**Conclusion:**
This skew toward higher ratings can impact collaborative filtering algorithms, as it introduces a **bias toward positive reinforcement**. To address this, normalization techniques or confidence-weighted scores can be considered. Understanding rating behavior helps design better loss functions and evaluation strategies for the final model.


```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Count number of ratings per movie
movie_ratings_count = ratings['movieId'].value_counts()

plt.figure(figsize=(10,6))
plot = sns.histplot(movie_ratings_count, bins=100, kde=True, color='skyblue')

# Add mean and median
plt.axvline(movie_ratings_count.mean(), color='red', linestyle='--', label='Mean')
plt.axvline(movie_ratings_count.median(), color='green', linestyle='--', label='Median')

# Log scale for better readability
plt.xscale('log')

# Title and labels
plt.title("Distribution of Ratings per Movie (Log Scale)")
plt.xlabel("Number of Ratings (log scale)")
plt.ylabel("Number of Movies")
plt.legend()
plt.grid(True)

# Annotate the peak bar
peak_bin = plot.patches[np.argmax([p.get_height() for p in plot.patches])]
peak_x = peak_bin.get_x() + peak_bin.get_width() / 2
peak_y = peak_bin.get_height()
plt.text(peak_x, peak_y + 500, f"Peak: ~{int(peak_y)} movies", ha='center', color='black')

plt.show()
```

### Distribution of Ratings per Movie

This plot illustrates the distribution of the number of ratings received by movies in the dataset. The x-axis is on a logarithmic scale to accommodate the long-tail nature of the data.

- Most movies have very few ratings, with the distribution being **heavily right-skewed**.
- A small number of popular movies receive a very high number of ratings.
- The **mean** (dashed red line) is much higher than the **median** (dashed green line), further confirming the skew.
- The **peak bin (~22,855 movies)** represents the majority of movies that received only a few ratings.

This insight is crucial for recommendation modeling, as it suggests the need for techniques that can handle data sparsity and popularity bias effectively.


```python
# Ratings per User Distribution
user_ratings_count = ratings.groupby('userId')['movieId'].count()
plt.figure(figsize=(10,6))
sns.histplot(user_ratings_count, bins=100, kde=True, color='lightcoral')
plt.title("Distribution of Ratings per User")
plt.xlabel("Number of Ratings")
plt.ylabel("Number of Users")
plt.xscale('log')  
plt.grid(True)
plt.show()
```

This plot displays the distribution of the number of ratings submitted by each user. The x-axis uses a logarithmic scale to better visualize the long-tail nature of the data.

- Most users have rated fewer than 100 movies, highlighting user sparsity.

- The plot is heavily right-skewed, indicating that:

- A small number of users are highly active, contributing thousands of ratings.

- A large portion of users are passive, rating only a handful of movies.

- This imbalance is typical in real-world recommendation systems and has several implications:

- It can bias recommendations toward active users.

- Models must be able to generalize from limited data to make effective predictions for users with very few ratings.

Understanding user behavior helps in selecting appropriate algorithms (e.g., matrix factorization with regularization, or hybrid methods) to handle cold-start users and data sparsity challenges.

```python
#Number of Ratings per Movie

ratings_per_movie = ratings.groupby('movieId')['rating'].count().sort_values(ascending=False)
plt.figure(figsize=(10, 6))
sns.histplot(ratings_per_movie, bins=50, kde=False)
plt.title('Distribution of Ratings per Movie')
plt.xlabel('Number of Ratings')
plt.ylabel('Number of Movies')
plt.grid(True)
plt.show()
```

```python
# Check for Class Imbalance in Ratings (Response Variable Distribution)

import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 5))
sns.countplot(
    x='rating',
    data=ratings,
    hue='rating',           #  Assign same column to hue
    palette='pastel',
    legend=False
)
plt.title('Distribution of Ratings')
plt.xlabel('Rating')
plt.ylabel('Count')
plt.grid(True)
plt.show()
```

###  Distribution of Ratings

This bar plot shows the distribution of movie ratings provided by users. From the chart, we observe that:

- Ratings are **skewed toward the higher end**, especially around **4.0**, which is the most frequent rating.
- Very few users rate movies below **2.0**, indicating a **positive bias** in the dataset.
- The ratings are in **half-point intervals** ranging from 0.5 to 5.0.

This class imbalance insight is important when building collaborative or hybrid recommender systems, as most users tend to give favorable ratings. This could influence model performance and may require **bias correction or weighting** techniques during training.

```python
ratings = pd.read_csv("../data/raw/rating.csv")

# Manually parse the timestamp using exact format
ratings['timestamp'] = pd.to_datetime(ratings['timestamp'], format="%Y-%m-%d %H:%M:%S", errors='coerce')

# Extract year
ratings['year'] = ratings['timestamp'].dt.year

# Confirm
print(" Fixed: ", ratings['timestamp'].notna().sum(), "valid timestamps")
print(ratings['year'].describe())
```

```python
# Create filtered DataFrame with valid years for boxplot
ratings_with_year = ratings.dropna(subset=['year', 'rating'])
ratings_with_year['year'] = ratings_with_year['year'].astype(int)
ratings_with_year = ratings_with_year[(ratings_with_year['year'] >= 1995) & (ratings_with_year['year'] <= 2015)]
```

```python
# plot 
plt.figure(figsize=(12, 6))
sns.boxplot(x='year', y='rating', data=ratings_with_year)
plt.xticks(rotation=45)
plt.title("Distribution of Ratings Over Time")
plt.grid(True)
plt.tight_layout()
plt.show()
```

###  Distribution of Ratings Over Time (1995–2015)

The boxplot above displays the distribution of user ratings across different years between 1995 and 2015.

####  Key Observations:
- **Stable Median:** The median rating remains close to 4.0 throughout the entire period, indicating generally favorable reviews.
- **Outliers:** A number of low-rating outliers (especially below 2.0) are present in each year, suggesting some movies consistently received poor feedback.
- **Slight Increase (2011–2014):** There is a slight increase in median and spread after 2010, possibly indicating a shift in user behavior or rating platform interface.
- **Consistent Spread:** The interquartile range (middle 50% of data) remains relatively consistent, showing stable variability in ratings.

This visualization confirms that while most ratings are positive, users do express a range of opinions. These yearly trends can guide time-based filtering or model features in the recommendation system.


```python
# Correlation Heatmap
corr_matrix = ratings[['userId', 'movieId', 'rating', 'year']].corr()

plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f", square=True)
plt.title("Correlation Heatmap")
plt.tight_layout()
plt.show()
```

###  Correlation Heatmap

This heatmap visualizes the **Pearson correlation coefficients** between numerical features: `userId`, `movieId`, `rating`, and `year`.

####  Key Observations:
- **Minimal correlation between `userId`, `movieId`, and `rating`**: All are close to zero, indicating these IDs are identifiers without direct numerical relationships.
- **Slight positive correlation between `movieId` and `year` (~0.46)**: Suggests newer movies (with higher IDs) were rated more in recent years.
- **No strong correlation with `rating`**: This implies user ratings do not vary strongly based on year or specific movie/user IDs, supporting the idea that user preferences are subjective.

These correlations guide feature engineering and modeling. Since most variables are uncorrelated, they likely contribute unique information to the recommendation algorithm.

```python

# Step 1: Merge ratings with movies (if not already done)
ratings_movies = pd.merge(ratings, movies[['movieId', 'title', 'genres']], on='movieId', how='left')

# Step 2: Replace (no genres listed) with real np.nan (DO NOT convert to string here)
ratings_movies['genres'] = ratings_movies['genres'].replace("(no genres listed)", np.nan)

# Step 3: Drop rows where genres is NaN
ratings_movies = ratings_movies.dropna(subset=['genres'])

# Step 4: Now safely split and explode
ratings_movies['genres'] = ratings_movies['genres'].str.split('|')
ratings_exploded = ratings_movies.explode('genres')

# Step 5: Sanity check
print(ratings_exploded['genres'].value_counts().head(10))
```

```python
#plot
plt.figure(figsize=(12, 6))
sns.countplot(y='genres', data=ratings_exploded, order=ratings_exploded['genres'].value_counts().index)
plt.title("Distribution of Ratings by Genre")
plt.xlabel("Number of Ratings")
plt.ylabel("Genre")
plt.tight_layout()
plt.show()
```

This plot shows the distribution of movie ratings across different genres after splitting multi-genre entries.

- Drama receives the highest number of ratings, followed by Comedy, Action, and Thriller.

- Less common genres like Film-Noir, Documentary, and Western have significantly fewer ratings.

- This reflects general viewer interest and content availability trends, suggesting genres like Drama and Comedy are more popular or widely available.

This analysis can guide:

- Feature weighting during modeling (e.g., giving genre-based embeddings or priors),

- Insights for personalization (e.g., genre-based filtering),

- Understanding of user behavior patterns and content diversity.


```python
#Average Rating per Genre
# Calculate average rating per genre
avg_rating_per_genre = ratings_exploded.groupby('genres')['rating'].mean().sort_values(ascending=False)

# Plot
plt.figure(figsize=(12, 6))
sns.barplot(x=avg_rating_per_genre.values, y=avg_rating_per_genre.index)
plt.title("Average Rating by Genre")
plt.xlabel("Average Rating")
plt.ylabel("Genre")
plt.tight_layout()
plt.show()
```

###  Average Rating by Genre

This horizontal bar chart illustrates the **average rating** given to movies across different genres.

- **Film-Noir, War, and Documentary** genres receive the highest average ratings, suggesting that although they may be less common, they are appreciated deeply by viewers.
- **Drama** and **Crime** also rank high, reflecting strong storytelling and emotional engagement.
- **Comedy**, **Children**, and **Horror** tend to receive relatively lower ratings on average, possibly due to wider variance in content quality or differing audience expectations.

Understanding the average rating by genre helps tailor recommendations toward **quality-focused suggestions**, not just popular ones. For instance:
- Recommender systems can **boost underrepresented but highly rated genres**.
- It helps identify genres that **polarize** viewers (high vs. low average ratings).

This analysis complements genre popularity by adding a **qualitative layer**—giving insight into not just *what users watch*, but *what they enjoy most*.


```python
# Ratings Over Time (Trend)
# Extract year from timestamp
ratings_exploded['year'] = pd.to_datetime(ratings_exploded['timestamp'], errors='coerce').dt.year

# Plot number of ratings by year
ratings_by_year = ratings_exploded.groupby('year').size()

plt.figure(figsize=(12, 6))
ratings_by_year.plot(kind='line', marker='o')
plt.title("Number of Ratings Over Time")
plt.xlabel("Year")
plt.ylabel("Number of Ratings")
plt.grid(True)
plt.tight_layout()
plt.show()
```

###  Number of Ratings Over Time

```python
# User Activity Distribution
# Count ratings per user
ratings_per_user = ratings_exploded['userId'].value_counts()

plt.figure(figsize=(10, 6))
sns.histplot(ratings_per_user, bins=50, kde=True)
plt.title("Distribution of Ratings per User")
plt.xlabel("Number of Ratings")
plt.ylabel("Number of Users")
plt.tight_layout()
plt.show()
```

###  Distribution of Ratings per User

This plot shows how many ratings each user has given across the dataset. Most users tend to rate only a few movies, with the number of users sharply declining as the number of ratings increases. 

The histogram is highly right-skewed, indicating a long tail of power users who contribute disproportionately more ratings. 

This kind of distribution is typical in recommender system datasets, where a small percentage of users are highly active while the majority provide only minimal feedback.

This insight is useful when designing collaborative filtering models, as special handling or regularization might be needed to manage such user activity imbalance.


##  Conclusion and Summary

In this exploratory data analysis (EDA), we examined the structure and behavior of the MovieLens dataset to extract insights that will guide our modeling phase. Key findings include:

- **Genre Distribution**: Drama, Comedy, and Action dominate in terms of rating volume, while Film-Noir and Documentary receive the highest average ratings.
- **Temporal Trends**: User activity peaked around 2000–2005 and gradually declined afterward.
- **User Behavior**: A small number of users contributed a large portion of ratings, indicating a skewed distribution.
- **Genre Preferences**: Genre popularity does not always align with rating quality, suggesting different user expectations across genres.

These insights are crucial for building a personalized movie recommendation system that takes into account user preferences, rating patterns, and genre behavior.

##  Next Steps: Building a Hybrid Recommendation System

Based on the insights from our exploratory data analysis, we plan to implement a **Hybrid Recommendation System** that combines the strengths of different techniques to provide more accurate and personalized suggestions.

###  Content-Based Filtering
We will use metadata such as genres, tags, and movie titles to compute similarity between items. 

This model will recommend movies similar to those a user has liked in the past, based on content attributes.

**Planned Features**:
- Genre encoding (multi-hot or embeddings)
- Tag-based similarity

###  Deep Learning-Based Recommendation
We will explore deep models to capture complex nonlinear patterns in user preferences.

**Approach**:
- Neural Collaborative Filtering (NCF)
- Embedding layers for users and items
- Dense layers to model interactions

###  Hybrid Model Strategy
The final recommendation system will combine the outputs from the above methods. Possible strategies include:
- Weighted hybrid (assigning scores from each model)
- Stacking (using a meta-model to combine predictions)
- Switching (choosing a method based on user profile)

---

##  Why This Approach?
- **Content-Based** ensures recommendations even for new users/items.
a- **Deep Learning** captures hidden, high-order patterns.
- **Hybridization** provides robustness and handles cold-start issues effectively.

This multi-model pipeline will help us design a more **scalable**, **accurate**, and **user-adaptive** movie recommendation engine.
